```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('ticks') # darkgrid, whitegrid, dark, white, ticks
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True)
```

```python
sns.set_style('ticks')
plt.figure(figsize=(10, 8))
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, size=4, aspect=2)
```

```python
sns.set_context('poster')
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.set_context('poster', font_scale=1.5)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='cool') # esquemas de cores --> https://matplotlib.org/examples/color/colormaps_reference.html
```
